# Libraries

```python
import pandas as pd
```

# Reading CSV

```python
# Read in the data from the csv file and store it in a dataframe and remove thr first 7 rows
df = pd.read_csv("../../Data/Ethnic_Groups/Ethnic_group_2011.csv", skiprows=7)
```

```python
df.head()
```

```python
df_melted = df.melt(id_vars='local authority: county / unitary (prior to April 2015)', 
                    var_name='Ethnic group', 
                    value_name='Observation')
df_melted
```

```python
df_melted = df_melted.rename(columns={'local authority: county / unitary (prior to April 2015)': 'Borough' })
df_melted.reset_index(drop=True, inplace=True)
df_melted.head(40)
```

# Dropping Null values and Duplicates

```python
df_melted.isna().sum()
```

```python
df_melted.dropna(inplace=True)
```

```python
df_melted.isna().sum()
```

```python
df_melted.duplicated().sum()
```

```python
mapping = {
    'White': 'White',
    'Mixed/multiple ethnic groups': 'Mixed',
    'Asian/Asian British': 'Asian',
    'Black/African/Caribbean/Black British': 'Black',
    'Other ethnic group': 'Other'
}

df_melted['Ethnic group'] = df_melted['Ethnic group'].replace(mapping)

df_melted
```

```python
df_melted['Ethnic group'].unique()
```

```python
df_melted
```

# Exporting CSV

```python
# export the cleaned data to a csv file
df_melted.to_csv('../../Data/Ethnic_Groups/Ethnic_group_2011_cleaned.csv', index=False)
```
