# Groupping the ethnicities into White, Black, Asian, Mixes, Other

```python
# Function for categorizing the ethnicities
def ethnicity_catecories(sentence):
    words = sentence.split()
    first_word = words[0]
    if '/' in first_word:
        first_word = first_word.split('/')[0]

    return first_word

ethnicity_catecories("White - English/Welsh/Scottish/Northern Irish/...")
```

```python
met_sas_2019_correct['Self-defined ethnicity'] = met_sas_2019_correct['Self-defined ethnicity'].apply(ethnicity_catecories)
```
